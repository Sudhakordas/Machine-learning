```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Stock_data = pd.read_csv('Google_Stock_Price_Train.csv', index_col = 'Date', parse_dates = True)
```

```python
Stock_data.info()
```

```python
Stock_data['Close'] = Stock_data['Close'].str.replace(',', '').astype(float)
```

```python
Stock_data['Volume'] = Stock_data['Volume'].str.replace(',', '').astype(float)
```

```python
Stock_data['Open'].plot(figsize = (16,5))
```

```python
Stock_data.rolling(window=30).mean()['Open'].plot()
```
